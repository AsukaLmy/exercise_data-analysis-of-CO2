we pick the dataset that contain In-situ CO2, flask CO2, and isotopic data sets taken at La Jolla Pier: Latitude 32.9°N Longitude 117.3°W Elevation 10m.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### import data

```python
df = pd.read_csv('daily_merge_co2_ljo.csv')

co2 = df[df['cri'] < 1][['Date3', 'co2']]
co2_1970 = co2[(co2['Date3']>1969)][['Date3', 'co2']]
co2_10_20 = co2[(co2['Date3'] < 2021) & (co2['Date3']>2009)][['Date3', 'co2']]

# transfer to numpy
co2np = co2.to_numpy()
print(co2np[:10,:])

```

```python
plt.figure(figsize=(10, 6))
plt.plot(co2_1970['Date3'], co2_1970['co2'], '-r', label='co2')
plt.title('co2')
plt.xlabel('date')
plt.ylabel('co2')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

## plotfit

```python
# line fit
coefficients_linear = np.polyfit(co2_1970['Date3'], co2_1970['co2'], 1)
polynomial = np.poly1d(coefficients_linear)
print(f"k = {coefficients_linear[0]}, b = {coefficients_linear[1]}")

# plot
plt.plot(co2_1970['Date3'], co2_1970['co2'], '-r',label='Data Points')
plt.plot(co2_1970['Date3'], polynomial(co2_1970['Date3']), '-b', label='Fitted Line')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
# line fit
coefficients_quad = np.polyfit(co2_1970['Date3'], co2_1970['co2'], 2)
polynomial = np.poly1d(coefficients_quad)
print(f"ax^2+bx+c: a = {coefficients_quad[0]}, b = {coefficients_quad[1]}, c = {coefficients_quad[2]}")

# plot
plt.plot(co2_1970['Date3'], co2_1970['co2'], '-r',label='Data Points')
plt.plot(co2_1970['Date3'], polynomial(co2_1970['Date3']), '-b', label='Fitted Line')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

From the fitting results, it can be observed that the annual variation of carbon dioxide concentration in the Hawaiian atmosphere roughly follows a quadratic curve distribution. Its rate of increase is accelerating over time. This is speculated to be due to the acceleration of modernization, with energy consumption increasing year by year, leading to a corresponding annual increase in the rate of carbon dioxide concentration.

## seasonal variations
## 季节性变化


In the dataset, we selected the changes in carbon dioxide content from 2010 to 2020. We can observe that while the overall carbon dioxide content in the Hawaiian atmosphere is increasing, it also shows periodic changes. Within a year, the carbon dioxide content in winter is usually higher than in summer. There are several reasons speculated for this: the strong photosynthesis of vegetation in summer leads to a significant reduction in carbon dioxide concentration; the concentrated heating demand in winter and the burning of fossil fuels result in an increase in carbon dioxide concentration.
